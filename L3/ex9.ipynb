### Excercise 13

Initialize random generator with 1

```python
import numpy as np
from numpy import random

random.seed(1)
```

a)

```python
X = random.normal(0,1,100)
print(X[:5])
```

b)

```python
eps = random.normal(0,0.25,100)
print(eps[:5])
```

c)

```python
Y = -1 + 0.5*X + eps
print(Y[:5])
```

len(Y) = 100

$\beta_1 = \frac{1}{2}$

$\beta_0 = \epsilon - 1$

d)

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

e)

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X.reshape((-1,1)), Y)

print(model.coef_[0], model.intercept_)
print(model.coef_[0] - 0.5, model.intercept_ - (-1))
```

f)

```python
Y_model = X * model.coef_[0] + model.intercept_
Y_reg = -1 + 0.5*X
plt.plot(X, Y_model, label='Model')
plt.plot(X, Y_reg, label='Population regression')
plt.scatter(X, Y, label='Real')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

g)

```python
X_2 = X**2
X_pol = np.stack((X, X_2), axis=-1)

import statsmodels.api as sm

X_pol = sm.add_constant(X_pol, prepend=True)
model = sm.OLS(Y, X_pol)
result = model.fit()
print(result.summary())
```
