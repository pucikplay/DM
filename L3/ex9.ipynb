### Excercise 14

a)

```python
rng = np.random.default_rng(10)
x1 = rng.uniform(0, 1, size=100)
x2 = 0.5 * x1 + rng.normal(size=100) / 10
y = 2 + 2 * x1 + 0.3 * x2 + rng.normal(size=100)
```

$y = 2x_1 + 0.3x_2 + 2$

b)

```python
plt.scatter(x1, x2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

c)

```python
x_cmb = np.stack((x1, x2), axis=-1)
x_cmb = sm.add_constant(x_cmb, prepend=True)
model = sm.OLS(y, x_cmb)
result = model.fit()
print(result.summary())
```

Can reject $H_0: \beta_1 = 0$, but not $H_0: \beta_2 = 0$

d)

```python
x = sm.add_constant(x1, prepend=True)
model = sm.OLS(y, x)
result = model.fit()
print(result.summary())
```

Can reject $H_0: \beta_1 = 0$

e)

```python
x = sm.add_constant(x2, prepend=True)
model = sm.OLS(y, x)
result = model.fit()
print(result.summary())
```

Can reject $H_0: \beta_2 = 0$

f)

No, they are not contradicting, as in c) $x_2$ didn't provide any "new information", but in e) it was the only source of it.

g)

```python
plt.scatter(x1, x2)
plt.scatter(0.1, 0.8)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()

plt.scatter(x1, y)
plt.scatter(0.1, 6)
plt.xlabel('x1')
plt.ylabel('y')
plt.show()

plt.scatter(x2, y)
plt.scatter(0.8, 6)
plt.xlabel('x2')
plt.ylabel('y')
plt.show()

x1 = np.concatenate([x1, [0.1]])
x2 = np.concatenate([x2, [0.8]])
y = np.concatenate([y, [6]])

x_cmb = np.stack((x1, x2), axis=-1)
x_cmb = sm.add_constant(x_cmb, prepend=True)
model = sm.OLS(y, x_cmb)
result = model.fit()
print(result.summary())
```

$x_2$ is now significant, $x_1$ isn't

```python
plt.scatter(result.get_influence().hat_matrix_diag[:-1], result.resid_pearson[:-1])
plt.scatter(result.get_influence().hat_matrix_diag[-1], result.resid_pearson[-1])
plt.plot([0,0.35],[2,2])
plt.plot([0,0.35],[-2,-2])
plt.xlabel('Leverage')
plt.ylabel('Residuals (standardised)')
```

```python
x = sm.add_constant(x1, prepend=True)
model = sm.OLS(y, x)
result = model.fit()
print(result.summary())
plt.scatter(result.get_influence().hat_matrix_diag[:-1], result.resid_pearson[:-1])
plt.scatter(result.get_influence().hat_matrix_diag[-1], result.resid_pearson[-1])
plt.plot([0.01,0.045],[2,2])
plt.plot([0.01,0.045],[-2,-2])
plt.xlabel('Leverage')
plt.ylabel('Residuals (standardised)')
```
